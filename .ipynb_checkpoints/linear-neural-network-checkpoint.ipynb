```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1,3,3],[1,4,3],[1,1,1],[1,0,2]])
Y = np.array([[1],[1],[-1],[-1]])
W = (np.random.random([3,1])-0.5)*2
print(W)
lr = 0.11
O=0
def update():
    global X,Y,W,lr
    O=np.dot(X,W)
    W_C=lr*(X.T.dot(Y-O))/int(X.shape[0])
    W = W + W_C
```

```python
for _ in range(100):
    update()
    x1 = [3,4]
    y1 = [3,3]
    x2 = [1,0]
    y2 = [1,2]
    k = -W[1]/W[2]
    d = -W[0]/W[2]
    print('k=',k)
    print('d=',d)
    xdata = (0,5)
    plt.figure()
    plt.plot(xdata,xdata*k+d,'r')
    plt.scatter(x1,y1,c='b')
    plt.scatter(x2,y2,c='y')
    plt.show()
    
```
